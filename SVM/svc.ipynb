## support vector classifier

```python
import pandas as pd
import numpy as numpy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## lets create a synthetic data points
from sklearn.datasets import  make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=12)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue = y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.25,random_state =10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel= 'linear')
```

```python
svc.fit(X_train,y_train)
```

```python
# predictions
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC (kernel='rbf')
rbf.fit(X_train, y_train)
# prediction
y_pred1 = rbf.predict(X_test)
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial = SVC (kernel='poly')
polynomial.fit(X_train, y_train)
# prediction
y_pred2 = polynomial.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
## hyperparameter tuning with GRID SEARCH CV
from sklearn.model_selection import GridSearchCV

## defining the parameters range

param_grid = {'C':[0.1,1,10,100,1000],
              'gamma' : [1,1.0,0.01,0.001,0.0001],
              'kernel' : ['rbf','linear']
}
```

```python
grid = GridSearchCV( estimator=SVC(), param_grid=param_grid, cv = 5, verbose= 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred3= grid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```
