```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
#feature engineering on Date_of_Journey

df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]

pd.concat([df['Date'],df['Month'],df['Year']],axis=1)
```

```python
df.head()
```

```python
df.info()
#still Date,Month,Year are object, need to convert them
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
#now we can drop Date_of_Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Now we should do the same step that we've done in the previous steps For Dep_Time
df['Dep_Time_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Time_Minute']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.info()#now we need to convert the Dep_Time_Hour and Dep_Time_Minute from object to int or float
```

```python
df['Dep_Time_Hour']=df['Dep_Time_Hour'].astype(int)
df['Dep_Time_Minute']=df['Dep_Time_Minute'].astype(int)
df.info()
```

```python
# Now we can remove the Dep_time permenantly
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
# now take the next column arrival_time
# First of all we need need to remove the unwanted text like '01:10 22 Mar' -> '01:10"
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_Time']
```

```python
df['Arrival_Time_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Time_Minute']=df['Arrival_Time'].str.split(':').str[1]
df.info()
```

```python
df['Arrival_Time_Hour']=df['Arrival_Time_Hour'].astype(int)
df['Arrival_Time_Minute']=df['Arrival_Time_Minute'].astype(int)
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
# now with Duratoin

df['Duration_Hour'] = df["Duration"].apply(lambda x: x.split('h')[0].strip())  # Extract hours
df['Duration_Minute'] = df["Duration"].apply(lambda x: x.split(' ')[1][:-1] if ' ' in x else '0')  # Extract minutes if available, else set to '0'

print(pd.concat([df['Duration'],df['Duration_Hour'],df['Duration_Minute']],axis=1))
df.head()
```

```python
# Now we can remove the Duration column
df.drop('Duration',axis=1,inplace=True)
```
