```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
encoder=OneHotEncoder()
enco_sex=encoder.fit_transform(df[['sex']]).toarray()
print(enco_sex)
```

```python
enco_sex_df=pd.DataFrame(enco_sex,columns=encoder.get_feature_names_out())
enco_sex_df
```

```python
print(df)
df.drop('sex',axis=1,inplace=True)
encoded_df=pd.concat([df,enco_sex_df],axis=1)
encoded_df
```

```python
enco_day=encoder.fit_transform(df[['day']]).toarray()
enco_day
```

```python
enco_day_df=pd.DataFrame(enco_day,columns=encoder.get_feature_names_out())
enco_day_df
```

```python
encoded_df.drop('day',axis=1,inplace=True)
```

```python
encoded_df=pd.concat([encoded_df,enco_day_df],axis=1)
encoded_df
```

```python
enco_smoker=encoder.fit_transform(df[['smoker']]).toarray()
enco_smoker
```

```python
enco_smoker_df=pd.DataFrame(enco_smoker,columns=encoder.get_feature_names_out())
print(enco_smoker_df)
```

```python
encoded_df.drop('smoker',axis=1,inplace=True)
encoded_df
```

```python
encoded_df=pd.concat([encoded_df,enco_smoker_df],axis=1)
encoded_df
```
